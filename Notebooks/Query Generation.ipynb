```python
!pip install transformers -q
!pip install sentence-transformers -q

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

## Reading The Data

```python
df = pd.read_csv('/content/links.csv')
df.head(5)
```

```python
# Converting Input to List Of Strings
text_list = df['text'].tolist()
n_gram_range = (1, 4)
stop_words = "english"

# Extract candidate words/phrases
count = CountVectorizer(ngram_range = n_gram_range, stop_words = stop_words).fit(text_list)
keywords = count.get_feature_names()
```

## Creating Word Embeddings

```python
topic_embeddings = model.encode(text_list)
keyword_embeddings = model.encode(keywords)

print(f'Length Of Topic Embeddings: {len(topic_embeddings)}\nLength Of Keyword Embeddings: {len(keyword_embeddings)}')
```

```python
top_n = 11000

distances = cosine_similarity(topic_embeddings, keyword_embeddings)
len(distances)
ex_keywords = [keywords[index] for index in distances.argsort()[0][-top_n:][::-1]] # most frequent to least

len(ex_keywords)
```

```python
query_df = pd.DataFrame(ex_keywords)
query_df.to_csv('query_df.csv')
```
